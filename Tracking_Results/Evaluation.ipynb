```
import matplotlib.pyplot as plt
import pandas as pd

# 1. Define the Data
data = {
    "Model": ["RegularSORT", "ByteTrack", "MeMOT", "ObjectCentricSORT", "StrongSORT"],
    "Total IDs": [32, 119, 380, 224, 244],
    "Avg Life (Frames)": [1947, 202, 67, 65, 27],
    "Max Life": [3546, 1088, 2041, 748, 351],
    "Gap Count": [30, 86, 123, 149, 174]
}

df = pd.DataFrame(data)

# 2. Setup the Plot
fig, ax = plt.subplots(figsize=(10, 4)) # Adjust width/height
ax.axis('off')
ax.axis('tight')

# 3. Create the Table
# cellLoc='right' aligns numbers to the right, 'left' for text
table = ax.table(cellText=df.values,
                 colLabels=df.columns,
                 cellLoc='right',
                 loc='center',
                 cellColours=[['#ffffff']*5]*5) # White background for cells

# 4. Styling
table.auto_set_font_size(False)
table.set_fontsize(12)
table.scale(1.2, 2) # Adjust spacing (width, height)

# Align the first column (Model Name) to the Left
for key, cell in table.get_celld().items():
    row, col = key
    if row == 0:
        # Header styling
        cell.set_text_props(weight='bold', color='black')
        cell.set_facecolor('#f2f2f2') # Light gray header
        cell.set_edgecolor('black')
    else:
        # Data styling
        cell.set_edgecolor('#dddddd') # Light gray borders
        if col == 0:
            cell.set_text_props(ha='left') # Left align model names
            cell.set_facecolor('#fafafa')  # Very light gray for model names

# 5. Save as JPEG
plt.title("Tracking Performance Comparison", pad=20, weight='bold', size=14)
plt.savefig('tracking_table.jpg', dpi=300, bbox_inches='tight', pad_inches=0.5)
plt.show()

print(" Image saved as 'tracking_table.jpg'. Check your file sidebar to download.")
```
